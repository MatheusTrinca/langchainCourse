### Text Loader

```python
from langchain_community.document_loaders import TextLoader

path = "./requirements.txt"

loader = TextLoader(path)

docs = loader.load()

print(docs)
print(len(docs))
print(docs[0].page_content)
```

### CSV Loader

### JSON Loader

```python
from langchain_community.document_loaders import JSONLoader

path = "./data/ecommerce.json"

loader = JSONLoader(path, jq_schema=".", text_content=False)

docs = loader.load()

print(docs)
print(len(docs))
print(docs[0].page_content)
```

### Structured File Loader

```python
from langchain_community.document_loaders import UnstructuredFileLoader, JSONLoader
from pathlib import Path

files = Path("data").glob("*")

docs = []

for file in files:
    docs.append(UnstructuredFileLoader(file).load())

print(docs)
print(len(docs))
print(docs[0].page_content)
```
